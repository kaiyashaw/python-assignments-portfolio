```python
# Q3 PART ONE
import numpy as np 
import matplotlib.pyplot as plt 

mean = 0
variance = 1
x = 0


def normal_density (mean, variance, x):
    f = (1/(np.sqrt(2 * np.pi * variance))) * (np.exp(-(x-mean)**2)/(2*variance))
    return f 

x = np.linspace (-10,10,1000)

#variables 1 
y_1 = normal_density(5,3,x)
plt.plot(x,y_1)

#variables 2 
y_2 = normal_density(1,3,x)
plt.plot(x,y_2)

#variables 3 
y_3 = normal_density(1,5,x)
plt.plot(x,y_3)

#plotting
plt.title("normal density")
plt.show()
```

```python
# Q4
import numpy as np
import matplotlib.pyplot as plt 

mean = 171
variance = 7.1**2
a = 162 
b = 190 

def normal_density (mean, variance, x):
    f = (1/(np.sqrt(2 * np.pi * variance))) * (np.exp(-(x-mean)**2)/(2*variance))
    return f 


#midpoint rule 
def integration (mean, variance, a, b, n = 1000000):
    delta_x = (b-a)/n
    points = np.linspace(a + delta_x/2, b - delta_x/2, n)
    f = normal_density (mean, variance, points) # function with midpoint points
    answer = np.sum( f * delta_x) #given by midpoint rule
    return (answer) 

def average_dose (mean, variance, a, b, n = 100000):
    dose = lambda x: 2.38*x**2
    delta_x = (b-a)/n
    points = np.linspace(a + delta_x/2, b - delta_x/2, n)
    f = normal_density (mean, variance, points)
    dose = dose(points)
    expected_dose = np.sum(dose*f*delta_x)
    return (expected_dose)

expected_dose = average_dose(mean,variance, a, b)

print("the estimated average dose is", expected_dose)
```
