```python
#Q4a

import sympy as sp
x,y = sp.symbols('x y')

#define the function
f = sp.exp(x) * sp.sin(y) + y**3

#partial derivative wrt x 
df_dx = sp.diff(f,x)

#partial derivative wrt y
df_dy = sp.diff(f,y)

print ("partial derivative wrt x =", df_dx)
print ("partial derivative wrt y =", df_dy)
```

```python
#Q4c

import sympy as sp 
x,y = sp.symbols('x y')

#define the function 
h = sp.log((x**2)+(y**2))

#second partial wrt x 
d2h_dx2 = sp.diff(h,x,x)
print ("second partial derivative wrt x =", d2h_dx2)

#second partial wrt y 
d2h_dy2 = sp.diff(h,y, y) 
print("second partial derivative wrt y =", d2h_dy2)

#second partial mixed 
d2h_dxdy = sp.diff(h,x,x,y,y)
print("second partial mixed derivative =", d2h_dxdy)

#the second partial mixed derivatives of symmetrical functions are always the same for "nice" functions because the second partial derivatives exist and are continuous,
#thus, for a nice function, the matric is symmetrical. This means that the order the function is differentiated in doesn't make a difference. 
```

```python
#Q4d

import sympy as sp 
import numpy as np 
import matplotlib.pyplot as plt 

x,y = sp.symbols ('x y')

#define the function 
j = x**3 - 3*x*y + y**3

#convert to numerical function 
j_func = sp.lambdify((x, y), j, 'numpy')

#grid of x and y and evaluate j 
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

#plot contour 
plt.contourf(X, Y, Z, levels=50, cmap='Spectral')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#Q4e

import sympy as sp 
import numpy as np 
import matplotlib.pyplot as plt 

x,y = sp.symbols ('x y')

#define function 
k = x**2 - y**2 #saddle function

#convert to numerical function 
k_func = sp.lambdify((x, y), k, 'numpy')

#grid of x and y and evaluate k 
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

#plot contour 
plt.contourf(X, Y, Z, levels=50, cmap='gnuplot2')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = x^2 - y^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
